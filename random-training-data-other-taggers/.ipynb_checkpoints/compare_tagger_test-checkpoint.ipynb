# This script compared the output of the taggers to the actual tags

```python
import matplotlib.pyplot as plt
```

```python
# open and read files and then split token and tags to save on list splitted_xy
with open("test_ittb.txt","r") as f:
    ittb = f.readlines()
splitted_ittb = [el.split() for el in ittb]

def compare_ittb(file):
    """this function is used to compare the output of the tagger with the ground truth of ittb"""
    num = 0
    correct = 0
    lookup_i = 0
    # open, read and split the content of the file you want to compare
    with open(file) as f:
        data = f.readlines()
        splitted_tagger = [el.split() for el in data]
    
    for i, e in enumerate(splitted_ittb):
        while lookup_i < len(splitted_tagger) and e[0] != splitted_tagger[lookup_i][0]:
            # Skip lines in splitted_tagger until e[0] matches
            lookup_i += 1
        
        if lookup_i < len(splitted_tagger):
            if e[1] == splitted_tagger[lookup_i][1]:
                correct += 1
            num += 1
            lookup_i += 1  # Move to the next line in splitted_tagger

    return (correct / num) * 100
```

```python
def plots(data, dataset):
    """this function creates plots displaying the accuracy over all treebanks. 
    it also returns the average accuracy."""
    labels = ['ITTB', 'LLCT', 'UDante', 'PROIEL', 'Perseus']
    # Create a bar plot for the single values
    plt.bar(labels, data)

    # Add labels
    plt.xlabel('Datasets')
    plt.ylabel('Accuracy')
    plt.title(f'Accuracy of the data tagged with {dataset}')
    
    
    # Add text labels for the values on top of the bars
    for i, v in enumerate(data):
        plt.text(i, v, str(round(v,2)), ha='center', va='bottom')

    # Show the plot or save it to a file
    plt.show()
    
    print(f"The average accuracy is {round(sum(data)/5,2)}")
```

```python
def map_tags(data):
    """this function is used to map the tags into upos tags"""
    map_tags = {
    # some of the tags do not have 1:1 corresponding tags and not all UPOS tags appear 
    # in this dictionary, for some i had to use my own defintions/intuition
    "ESSE:IND": "VERB",
    "ESSE:SUB": "VERB",
    "ESSE:INF": "VERB",
    "V:IND" : "VERB",
    "V:SUB": "VERB",
    "V:INF": "VERB",
    "V:GER": "VERB",
    "V:GED": "VERB",
    "V:PTC:nom": "VERB",
    "V:PTC:acc": "VERB",
    "V:PTC:abl": "VERB",
    "V:PTC": "VERB",
    "V:SUP:acc": "VERB",
    "V:SUP:abl": "VERB",
    "V:IMP": "VERB",
    "PRON":"PRON",
    "REL":"PRON",
    "POSS":"PRON",
    "DIMOS":"PRON",
    "INDEF":"PRON",
    "N:nom":"NOUN",
    "N:NOM":"NOUN",
    "N:dat":"NOUN",
    "N:gen":"NOUN",
    "N:loc":"NOUN",
    "N:acc":"NOUN",
    "N:abl":"NOUN",
    "N:voc":"NOUN",
    "ADJ:NUM":"NUM",
    "CC":"CCONJ",
    "CS":"SCONJ",
    "NPR":"PROPN",
    "ADJ":"ADJ",
    "ADJ:COM":"ADJ",
    "ADJ:SUP":"ADJ",
    "ADJ:abl":"ADJ",
    "ADV":"ADV",
    "PREP":"ADP",
    "INT":"INTJ",
    "ABBR":"X",
    "EXCL":"INTJ",
    "FW":"X",
    "SENT":"PUNCT",
    "PUN":"PUNCT",
    "SYM":"SYM",
    "CLI":"CCONJ",
    "DET":"DET",
    "ENCL":"CCONJ",
    }
    
    cleaned = []
    with open(data,"r") as f:
        content = f.readlines()
        splitted = [el.split() for el in content]
        for i,e in enumerate(splitted):
            if map_tags.get(e[1]) == None:
                print(e)
            cleaned.append([e[0],map_tags.get(e[1])])

    with open(data,"w") as f:
        for e in cleaned:
            row = e[0] + "\t"+e[1]+"\n"
            f.write(row)
            

```

```python
map_tags("test_sets/treetagger_ittb.txt")
map_tags("test_sets/treetagger_llct.txt")
map_tags("test_sets/treetagger_udante.txt")
map_tags("test_sets/treetagger_proiel.txt")
# i manually changed the line in the proiel file: quam ob rem	CS	quam ob rem
# into this: 
"""
quam    PRON    quam
ob  PREP     ob
rem N:NOM   res
"""
map_tags("test_sets/treetagger_perseus.txt")
```
