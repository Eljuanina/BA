```python
def calculate_tag_percentage_from_files(file_list, tag_position=1):
    """this function returns the tag distribution. 
    it takes as input a file list and counts the occurences of each tag within the files"""
    recognized_tags = {
        'ADJ', 'ADP', 'PUNCT', 'ADV', 'AUX', 'SYM',
        'INTJ', 'CCONJ', 'X', 'NOUN', 'DET', 'PROPN',
        'NUM', 'VERB', 'PART', 'PRON', 'SCONJ'
    }

    # Placeholder for tag counts
    tag_counts = {tag: 0 for tag in recognized_tags}
    other_count = 0
    total_count = 0

    for file_name in file_list:
        with open(file_name, 'r') as file:
            # Read the content of the file and process tags
            lines = file.readlines()
            for line in lines:
                # Split the line and get the tag at the specified position
                data = line.strip().split()
                if len(data) > tag_position:
                    tag = data[tag_position]
                    if tag in recognized_tags:
                        tag_counts[tag] += 1
                    else:
                        other_count += 1
                    total_count += 1

    # Calculate percentages
    tag_percentages = {tag: (count / total_count) * 100 for tag, count in tag_counts.items()}
    tag_percentages['other'] = (other_count / total_count) * 100

    return tag_percentages
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_grouped_bar(tag_percentages1, tag_percentages2, tag_percentages3, tag_percentages4, tag_percentages5):
    """this function creates a grouped bar plot from the percentage dictionaries"""
    recognized_tags = {
        'ADJ', 'ADP', 'PUNCT', 'ADV', 'AUX', 'SYM',
        'INTJ', 'CCONJ', 'X', 'NOUN', 'DET', 'PROPN',
        'NUM', 'VERB', 'PART', 'PRON', 'SCONJ'
    }

    tags = list(recognized_tags) + ['other']
    tags = sorted(tags)  # Sort for consistency

    percentages1 = [tag_percentages1[tag] for tag in tags]
    percentages2 = [tag_percentages2[tag] for tag in tags]
    percentages3 = [tag_percentages3[tag] for tag in tags]
    percentages4 = [tag_percentages4[tag] for tag in tags]
    percentages5 = [tag_percentages5[tag] for tag in tags]

    bar_width = 0.15
    index = np.arange(len(tags))

    fig, ax = plt.subplots(figsize=(12, 8))

    bars1 = plt.bar(index, percentages1, bar_width, label='CLTK')
    bars2 = plt.bar(index + bar_width, percentages2, bar_width, label='LatinCy')
    bars3 = plt.bar(index + 2 * bar_width, percentages3, bar_width, label='RDRPOSTagger')
    bars4 = plt.bar(index + 3 * bar_width, percentages4, bar_width, label='UDPipe')
    bars5 = plt.bar(index + 4 * bar_width, percentages5, bar_width, label='TreeTagger')

    plt.xlabel('Tags')
    plt.ylabel('Percentage')
    plt.title('Percentage of Tags')
    plt.xticks(index + 2 * bar_width, tags)
    plt.legend()

    plt.tight_layout()
    plt.savefig("tag_dist_tb_pretrained.png")
    plt.show()

plot_grouped_bar(tag_percentages_cltk, tag_percentages_latincy, tag_percentages_rdrpostagger, tag_percentages_udpipe, tag_percentages_treetagger)
```

```python
def calculate_tag_percentage_from_file(file_path, tag_position=1):
    """this function creates a grouped bar plot from the given filenames"""
    recognized_tags = {
        'ADJ', 'ADP', 'PUNCT', 'ADV', 'AUX', 'SYM',
        'INTJ', 'CCONJ', 'X', 'NOUN', 'DET', 'PROPN',
        'NUM', 'VERB', 'PART', 'PRON', 'SCONJ'
    }

    # Placeholder for tag counts
    tag_counts = {tag: 0 for tag in recognized_tags}
    other_count = 0
    total_count = 0

    with open(file_path, 'r') as file:
        # Read the content of the file and process tags
        lines = file.readlines()
        for line in lines:
            # Split the line and get the tag at the specified position
            data = line.strip().split()
            if len(data) > tag_position:
                tag = data[tag_position]
                if tag in recognized_tags:
                    tag_counts[tag] += 1
                else:
                    other_count += 1
                total_count += 1

    # Calculate percentages
    tag_percentages = {tag: (count / total_count) * 100 for tag, count in tag_counts.items()}
    tag_percentages['other'] = (other_count / total_count) * 100

    return tag_percentages

# List of models and datasets
models = ['50', '100', '200','500','1000','2000','5000','10000']  # Replace with your model names
datasets = ['ittb', 'llct', 'udante', 'proiel', 'perseus']  # Replace with your dataset names

percentages = []
for model in models:
    mod_percent = []
    for dataset in datasets:
        file_path = f"../GPT/output_train{model}-{dataset}-corr.txt"
        tag_percentages = calculate_tag_percentage_from_file(file_path)
        mod_percent.append(tag_percentages)
    percentages.append(mod_percent)

```
