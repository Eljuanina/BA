# Perseus

```python
# we take every perseus file except for the files containing
# Caesar Commentarii de Bello Gallico (2.1-2.3; 2.5; 2.7; 2.9; 2.14-2.18; 2.32-2.33)
# or Jerome  Vulgata
```

```python
filenames = ["phi1351.phi005.perseus-lat1.tb.xml","phi1348.abo012.perseus-lat1.tb.xml",
             "phi1221.phi007.perseus-lat1.tb.xml","phi0975.phi001.perseus-lat1.tb.xml",
             "phi0972.phi001.perseus-lat1.xml","phi0959.phi006.perseus-lat1.tb.xml",
             "phi0690.phi003.perseus-lat1.tb.xml","phi0631.phi001.perseus-lat1.tb.xml",
             "phi0620.phi001.perseus-lat1.tb.xml","phi0474.phi013.perseus-lat1.tb.xml"]
```

```python
# Initialize lists to store extracted data
sentence_num = []
word_ids = []
words = []
tags = []
num = 1
# Iterate through the files
for filename in filenames:
    # Load and parse the XML file
    with open(filename, "r", encoding="utf-8") as file:
        xml_data = file.read()
        root = ET.fromstring(xml_data)

        # Iterate through <sentence> elements
        for sentence_elem in root.findall(".//sentence"):
            # Iterate through <word> elements within each <sentence>
            for word_elem in sentence_elem.findall(".//word"):
                word_id = word_elem.get("id")
                word = word_elem.get("form")
                tag = word_elem.get("postag")
                # Append data to lists
                sentence_num.append(num)
                word_ids.append(word_id)
                words.append(word)
                tags.append(tag)
            num += 1
```

```python
# Create a DataFrame
data = {
    "s": sentence_num,
    "id": word_ids,
    "token": words,
    "pos": tags
}

df_perseus = pd.DataFrame(data)

df_perseus
```

```python
# map the tags to upos
```

```python
# first part: part of speech tag

# no difference of SCONJ and CCONJ here
mapped={
    "n" : "NOUN",
    "v" : "VERB",
    "a" : "ADJ",
    "d" : "ADV",
    "c" : "CONJ",
    "r" : "ADP",
    "p" : "PRON",
    "m" : "NUM",
    "i" : "INTJ",
    "e" : "X",
    "u" : "PUNCT",
    "None": "None",
    "-" : "None"
}

def extract_first_letter(pos):
    """This function will return the first letter, that means in that case the pos tag"""
    if pos != None and not(len(pos)==0):
        return pos[0]
    else:
        return "None"


def map_postags(df):
    """this function maps the tags of the 'pos' column into upos tags"""
    df["first_pos"] = df["pos"].apply(extract_first_letter)
    df["upos"] = df["first_pos"].map(mapped)
    return df
        
def clean_none(df):
    """this function removes all 'None' values from the df"""
    df = df[df['upos'] != 'None']
    return df


mapped_upos_df_perseus =  clean_none(map_postags(df_perseus))
mapped_upos_df_perseus
```

```python
def extract_from_df(df, num):
    """this function is used to extract random samples from a df using random seed"""
    # Get a list of all unique values in the "s" column
    all_values_s = df["s"].unique().tolist()
    
    # Seed the random number generator for reproducibility
    random.seed(4)
    
    # Select num random elements from the list of all_values_s
    random_sentences = random.sample(all_values_s, num)
    
    # Filter the DataFrame to include rows where the "s" column matches the random sentences
    df = df[df["s"].isin(random_sentences)]

    return df
```

```python
# write in jsonl for fine tuning format
def save_to_jsonl(file, data):
    """this function is used to save the data in jsonl format"""
    with open(file, 'w') as f:
        with open(data+".txt", "r") as d:
            sentence = ["'"]
            tagged = []
            system_prompt = '"You are a Latin linguist and part-of-speech tagging expert. You are using UPOS (universal part of speech tags). UPOS tags are ADJ,ADP,ADV,AUX,CCONJ,DET,INTJ,NOUN,NUM,PART,PRON,PROPN,PUNCT,SCONJ,SYM,VERB and X. X stands for other."'
            old_s = []
            last_tok = ""
            last_tag = ""
            for row in d:
                splitted = row.split()
                s,token, tag = splitted[0].strip(), splitted[1].strip(), splitted[2].strip()
                if len(old_s)>0 and old_s[-1]== s:
                    sentence.append(token)
                    tagged_row = token + '\\\\t' + tag + '\\\\n'  # Use "\\n" to print "\\n" literally
                    tagged.append(tagged_row)
                    last_tok = token
                    last_tag = tag
            
                else:
                    old_s.append(s)
                    sentence.append(last_tok)
                    sentence.append("'")
                    sent = ' '.join(sentence)
                    tagged_row = last_tok + '\\\\t' + last_tag + '\\\\n'
                    tagged.append(tagged_row)
                    completion = ' '.join(tagged)

                    prompt = '"Return the UPOS tag for the tokens of the sentence: ' + sent + 'Return the tags in the format TOKEN\\tTag\\n. Only output the token and the tag (no explanations, no translations, no additional text)."'

                    out = '{"messages":[{"role":"system","content":' + system_prompt + '},{"role":"user","content":' + prompt + '},{"role":"assistant","content":"' + completion + '"}]}\n'
                    f.write(out)
                    sentence = ["'"]
                    tagged = []
                    
                    sentence.append(token)
                    tagged_row = token + '\\\\t' + tag + '\\\\n'  # Use "\\n" to print "\\n" literally
                    tagged.append(tagged_row)
                    
```
