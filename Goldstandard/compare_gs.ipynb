```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import re
import numpy as np
```

```python
# remove punct from gs
def remove_punct(file):
    """this function removes all the punctuation symbols from the input file"""
    # define punctuation
    puncts = ",;.:-_!?'\"{([])}"
    # open the file gives as parameter
    with open(file) as file:
        # read the content
        content = file.readlines()
        # and split it (separate token and tags)
        splitted = [el.split() for el in content]
        # define an empty list to store the cleaned elements
        cleaned = []
        for element in splitted:
            # paramteter to store the current token char by char (cleaned)
            token = ""
            # for char in the token
            for char in element[0]:
                # if the char is not in the defined punctuation
                if char not in puncts:
                    # then add it to the variable
                    token += char
            # if token not empty when cleaned from punctuation
            if token == "":
                # continue
                continue
            # if not empty, then append it to the list with its tag
            else:
                cleaned.append([token,element[1]])
        return cleaned
```

```python
# Remove punctuation from first goldstandard
cleaned = remove_punct("Goldstandard_elina.txt")
labels1 = [el[1] for el in cleaned]

# Remove punctuation from second goldstandard
cleaned2 = remove_punct("Goldstandard_urs.txt")
labels2 = [el[1] for el in cleaned2]
```

```python
# Calculate Cohen's Kappa score
kappa_score = cohen_kappa_score(labels1, labels2)
print(f"Cohen's Kappa Score: {kappa_score:.2f}")
```

```python
def show_tag_distribution(dictionary,name):
    """this function shows the tag distribution of a input dictionary"""
    # Extract labels and values from the dictionary
    labels = list(dictionary.keys())
    values = list(dictionary.values())

    # Create a bar plot
    plt.figure(figsize=(6, 4))  # Set the figure size
    plt.bar(labels, values, color='teal')
    plt.xlabel('Part of Speech')
    plt.ylabel('Percentage')
    plt.title(f'Percentage of Part of Speech Tags in {name}')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

    # Display the plot
    plt.tight_layout()
    plt.savefig(f"{name}_tag_dist.png")
    plt.show()
```

# Function to plot confusion matrix

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def confusion(actual, predicted, name):
    """this function returns a confusion matrix with the upos tags from an input df compared to gold standard"""
    upos = ["VERB", "ADV", "ADJ", "NOUN", "PROPN", "PRON", "DET", "ADP", "SCONJ", "AUX", "NUM", "CCONJ", "PART", "X", "INTJ", "SYM"]

    # Calculate the confusion matrix
    cm = confusion_matrix(actual, predicted, labels=upos)

    # Calculate row sums
    row_sums = cm.sum(axis=1)

    # Calculate the confusion matrix percentages, handling division by zero
    cm_percentage = cm.astype("float") / np.maximum(row_sums, 1)[:, np.newaxis]

    # Create a heatmap to visualize the confusion matrix percentages
    plt.figure(figsize=(10, 8))
    ax = sns.heatmap(cm_percentage, annot=True, fmt=".2f", cmap="GnBu", xticklabels=upos, yticklabels=upos)

    # Change the font size of the tick labels
    ax.tick_params(axis='both', which='both', labelsize=8)  # You can adjust the labelsize as needed

    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{name} Confusion Matrix")
    plt.tight_layout()

    plt.savefig(f"{name}_confusion_matrix.png")  # Save the figure to a file
    plt.show()
```

```python
def remove_none(df):
    """this function is used to remove all 'None' from the df and replaces it with 'X'"""
    for i, row in df.iterrows():
        if row["tag"] is None or row["tag"] == "None":
            df.at[i, "tag"] = "X"
            
    return df
```

```python
def calculate_tag_percentage(dataframe, column_name):
    """return the tag distribution for a dataframe"""
    # Count occurrences of each tag
    tag_counts = dataframe[column_name].value_counts()

    # Calculate percentages
    total_elements = len(dataframe)
    tag_percentages = {tag: (count / total_elements) * 100 for tag, count in tag_counts.items()}
    
    return tag_percentages
```

```python
# save cleaned version of gs into files
with open("cleaned_gs1.txt","w") as o:
    for line in content:
        puncts = ",:;._-'\"!?(){}[ ]\t\n"
        splitted = line.split()
        if splitted[0].strip() not in puncts:
            tok = ""
            for char in splitted[0]:
                if char not in puncts:
                    tok += char
            if len(tok) > 0:
                row = tok+"\t"+splitted[1]+"\n"
                o.write(row)
```

```python
from collections import Counter

def calculate_tag_percentage_from_list(tag_list):
    """calculate the tag distribution from a list"""
    total_elements = len(tag_list)

    # Count occurrences of each tag in the list
    tag_counts = Counter(tag_list)

    # Calculate percentages for each tag in the list
    tag_percentages = {tag: (count / total_elements) * 100 for tag, count in tag_counts.items()}

    return tag_percentages
```

```python
def transform_dicts(dictionary):
    """this function transforms the dictionaries so that only upos tags and one columns named 'other' are contained"""
    recognized_tags = {
        'ADJ', 'ADP', 'PUNCT', 'ADV', 'AUX', 'SYM',
        'INTJ', 'CCONJ', 'X', 'NOUN', 'DET', 'PROPN',
        'NUM', 'VERB', 'PART', 'PRON', 'SCONJ'
    }

    new_dict = {tag: 0 for tag in recognized_tags}
    other_sum = 0

    for tag, value in dictionary.items():
        if tag in recognized_tags:
            new_dict[tag] = value
        else:
            other_sum += value

    
    new_dict['other'] = other_sum

    return new_dict
```

```python
def plot_grouped_bar_multiple(tag_percentage_dicts, names):
    """this function returns a grouped bar plot for taggers on the bullinger data. 
    as input it takes dictionaries containing the distribtuion of tags for certain taggers and their names"""
    recognized_tags = {
        'ADJ', 'ADP', 'PUNCT', 'ADV', 'AUX', 'SYM',
        'INTJ', 'CCONJ', 'X', 'NOUN', 'DET', 'PROPN',
        'NUM', 'VERB', 'PART', 'PRON', 'SCONJ'
    }

    # Plotting
    num_sources = len(tag_percentage_dicts)
    fig, ax = plt.subplots(figsize=(10, 6))

    bar_width = 0.15
    x = range(len(recognized_tags) + 1)  # 1 added for 'other'

    for i, tag_percentage_dict in enumerate(tag_percentage_dicts):
        percentages = [tag_percentage_dict[tag] if tag in tag_percentage_dict else 0 for tag in recognized_tags]
        other_value = tag_percentage_dict.get('other', 0)
        percentages.append(other_value)
        ax.bar([pos + bar_width * i for pos in x], percentages, width=bar_width, label=names[i] if i < len(names) else f'Source {i+1}')
    ax.set_xticks([pos + 0.5 * bar_width * (num_sources - 1) for pos in x])
    ax.set_xticklabels(list(recognized_tags) + ['other'])
    ax.set_xlabel('Tags')
    ax.set_ylabel('Percentage')
    ax.set_title('Percentage of Tags')
    ax.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig("bullinger_tagdist.png")
    plt.show()


tag_percentage_dicts = [
    transform_dicts(calculate_tag_percentage(df_gs1, "tag")),
    transform_dicts(calculate_tag_percentage(df_latincy, "tag")),
    transform_dicts(calculate_tag_percentage(df_cltk, "tag")),
    transform_dicts(calculate_tag_percentage(df_udpipe, "tag")),
    transform_dicts(calculate_tag_percentage(df_rdrpostagger, "tag")),
    transform_dicts(calculate_tag_percentage(df_treetagger, "UPOS"))
    
]


names = ['GS', 'LatinCy', 'CLTK', 'UDPipe','RDRPOSTagger', 'TreeTagger']

plot_grouped_bar_multiple(tag_percentage_dicts, names)
```
