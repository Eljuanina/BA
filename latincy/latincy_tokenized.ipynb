# LatinCy on Bullinger Data
### This script was used to apply LatinCy POS tagger on the Bullinger data using the large latin model

```python
import os
import spacy
nlp = spacy.load('la_core_web_lg')
```

```python
# go one directory level up to the parent
os.chdir("..")
# open and read the bullinger samples
with open("samples/tokenized.txt") as f:
    content = f.read().strip()
# analyze the content of the file
doc = nlp(content)
```

```python
# print the first example
print(f'{doc[0].text}, {doc[0].norm_}, {doc[0].lemma_}, {doc[0].pos_}')
```

```python
# go back into latincy folder
os.chdir("latincy")
```

```python
# write output token, pos into files
with open("latincy_tokenized.txt", "w") as out:
    for token in doc:
        # if not just newline character on line print to file with pos tag
        if token.text != "\n":
            line = token.text.strip()+"\t"+token.pos_+"\n"
            print(line)
            out.write(line)
```
