# Compare the results

```python
def compare(file, reference):
    correct = 0
    total = 0
    
    # open and read content of corrected tagged file
    with open(file,"r") as f:
        content = f.readlines()
    
    # open and read correct solution from reference
    with open(reference, "r") as r:
        ref = r.readlines()
        
    for i,e in enumerate(content):
        line = e.split()
        token, tag = line[0].strip(), line[-1].strip()
        row = ref[i].split()
        if token == row[0]:
            total += 1
            # i compare the lowercase versions of the tags because then gpt would have assigned it correctly,
            # only the format would have been wrong
            if tag.lower() == row[-1].lower():
                correct += 1
        else:
            print(i+1)
            print(e)
            print(row)
         
    return round((correct/total)*100,2)
```
